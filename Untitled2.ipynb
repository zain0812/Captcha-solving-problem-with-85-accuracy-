```python
import numpy as np
import cv2
import os
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

#load training images and labels
def load_training_images_and_labels(folder_path):
    images = []
    labels = []

    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg"):
            image_path = os.path.join(folder_path, filename)
            # Load image and resize to 64x64
            image = cv2.imread(image_path)
            image = cv2.resize(image, (64, 64))  # Resize the image
            image = image / 255.0  # Normalize the image

            # Extract label from filename for trining purpos
            label = filename.split('-')[-1].split('.')[0]  # Get the CAPTCHA solved text
            label = [int(digit) for digit in label]  # Convert to a list of digits

            images.append(image)
            labels.append(label)

    images = np.array(images)
    labels = np.array(labels)

    return images, labels

# Load the training data
train_folder = '/content/Training'
train_images, train_labels = load_training_images_and_labels(train_folder)

# padding labls to have same length
train_labels = tf.keras.preprocessing.sequence.pad_sequences(train_labels, padding='post', maxlen=6)

# One-hot encode the labels for each digit
train_labels = np.array([to_categorical(label, num_classes=10) for label in train_labels])

# Verify the shapes of the images and labels
print(f"train_images shape: {train_images.shape}")
print(f"train_labels shape: {train_labels.shape}")

# Define the CNN model
def build_model():
    input_img = Input(shape=(64, 64, 3))

    x = Conv2D(32, (3, 3), activation='relu')(input_img)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)

    # Output for each digit (6 digits in total)
    outputs = [Dense(10, activation='softmax', name=f'digit_{i}')(x) for i in range(6)]

    model = Model(inputs=input_img, outputs=outputs)

    # Compile the model with separate metrics for each output
    model.compile(
        optimizer=Adam(),
        loss='categorical_crossentropy',
        metrics=['accuracy'] * 6  # Accuracy for each of the 6 outputs
    )
    return model

# Buildong and compiling the model
model = build_model()
model.summary()

# Traing the model
model.fit(
    train_images,
    [train_labels[:, i] for i in range(6)],  # Labels split by digit
    epochs=20,
    batch_size=32
)
```

```python
import os
import numpy as np
import cv2

def load_images_and_labels(folder_path, img_size=(64, 64)):
    images = []
    labels = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg'):
            filepath = os.path.join(folder_path, filename)
            # Load and preprocess the image
            image = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)  # Load as grayscale
            image = cv2.resize(image, img_size) / 255.0  # Resize and normalize
            # Extract label from filename
            label = filename.split('-')[-1].split('.')[0]
            label = [int(digit) for digit in label]  # Convert label to a list of digits

            images.append(image)
            labels.append(label)
    images = np.array(images).reshape(-1, img_size[0], img_size[1], 1)  # Add channel dimension
    labels = np.array(labels)
    return images, labels
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

# Function to load testing images and labels with noise removal and resizing
def load_testing_images_and_labels(folder_path):
    images = []
    labels = []

    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg"):
            image_path = os.path.join(folder_path, filename)
            # Load image
            image = cv2.imread(image_path)

            # Greyscalling
            gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            # Gaussian blur to remove noise
            blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

            # Resizeing image (64x64)
            resized_image = cv2.resize(blurred_image, (64, 64))  # Resize to match model input size

            # Normalizeing image
            resized_image = resized_image / 255.0  # Normalize to [0, 1]

            # Convert grayscale to 3 channels by duplicating the grayscale values
            resized_image = np.stack([resized_image] * 3, axis=-1)  # Add channel dimension

            # Append the processed image
            images.append(resized_image)

            # For testing, labels are not available, so we set a placeholder value (-1)
            labels.append(-1)  # Testing images have unknown CAPTCHA labels

    images = np.array(images)
    labels = np.array(labels)
    return images, labels

# Load the testing data
test_folder = '/content/Testing New'
if not os.path.exists(test_folder):
    print(f"Error: The folder '{test_folder}' does not exist.")
else:
    print(f"The folder '{test_folder}' exists. Loading images...")

test_images, test_labels = load_testing_images_and_labels(test_folder)
print(f"test_images shape: {test_images.shape}")

# Predict and display a random testing image
def predict_and_display_random_image():
    # Ensure model and test_images are available
    if len(test_images) == 0:
        print("No test images available.")
        return

    # Select a random image from the test set
    random_index = random.randint(0, len(test_images) - 1)
    test_image = test_images[random_index]

    # Predict the CAPTCHA
    prediction = model.predict(np.expand_dims(test_image, axis=0))

    # Get the predicted labels for all digits (6 digits)
    predicted_label = ''.join(str(np.argmax(digit)) for digit in prediction)  # Combine predicted digits

    # Display the image and predicted text
    plt.imshow(test_image)
    plt.axis('off')
    plt.title(f"Predicted CAPTCHA: {predicted_label}")
    plt.show()

# Calling the output finction
predict_and_display_random_image()
```
